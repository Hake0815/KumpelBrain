```python
import sys
import torch

sys.path.insert(0, "../../cpp/build")
import kumpel_embedding
import positional_embedding

SEED = 42

# Set Python seed
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

torch.use_deterministic_algorithms(True)

# Create input
x = torch.randn(2, 10, 512)

# Python model
py_model = positional_embedding.PositionalEmbedding(512,0.9)
py_model.eval()

# C++ model
cpp_model = kumpel_embedding.PositionalEmbedding(512, 0.9, 5000)
cpp_model.eval()

# Forward
y_py = py_model(x)
y_cpp = cpp_model.forward(x)

# Compare
print(torch.allclose(y_py, y_cpp, atol=1e-6))
```
