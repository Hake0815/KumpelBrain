```python
import torch
import torch.nn as nn
import math
```

```python
max_len = 512
d_model = 8
position = torch.arange(max_len).unsqueeze(1)
div_term = torch.exp(torch.arange(1, d_model + 1) * (-math.log(10000.0) / d_model))
pe = torch.zeros(1, max_len, d_model)
pe[0, :, 0::2] = torch.sin(position * div_term[0::2])
pe[0, :, 1::2] = torch.cos(position * div_term[1::2])
```

```python
batch_size = 3
seq_len = 2
x=torch.rand(batch_size, seq_len, d_model)
print(x)
print(x.size(1))
print(pe[:, :x.size(1), :].size())
print(pe[:, :x.size(1), :])

print((x+pe[:, :x.size(1), :]).size())
print((x+pe[:, :x.size(1), :]))

```
